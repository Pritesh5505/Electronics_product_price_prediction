```python
import requests
from bs4 import BeautifulSoup
base_url = "https://www.flipkart.com"
import pandas as pd 
import numpy as np
import time 
import re
from fake_useragent import UserAgent
```

```python
ua = UserAgent()
header = {'user-Agent' : ua.random}
```

```python
productlinks = []

for i in range(1,42):
    try:
        r = requests.get(f'https://www.flipkart.com/search?q=TV&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}',headers = header)
        #print(i)
        soup = BeautifulSoup(r.content, 'lxml')
        productlist = soup.find('div',class_="_1YokD2 _3Mn1Gg")
        #print(productlist)
        for item in productlist.find_all('div', class_= "_1AtVbE col-12-12"):
            for link in item.find_all('a',href = True):
                productlinks.append(base_url + link['href'])
        #len(productlinks)
        k=12
        productlinks = productlinks[: len(productlinks)-k]
        #len(productlinks)
    except Exception as e:
        print(i)
        print("error has occured")
    time.sleep(np.random.randint(1,3))
```

```python
tv = []
rem_tv = []
for link in productlinks:
    try:
        
        r = requests.get(link,headers= header)
        print(r)
        soup = BeautifulSoup(r.content,'lxml')
        brand = soup.find('span', class_ = "B_NuCI").text
        #print(brand)
        soup1 = soup.find_all("tr", class_="_1s_Smc row")
        #print(soup1)
        model_name = display_size = os = pitcure_quality = Smart_Tv = Sound_Technology = Refresh_Rate = Screen_Type = HDMI = USB = "NA"
        for row in soup1:
            if row.find("td", string = "Model Name"):
                model_name = row.find('li', class_ = "_21lJbe").text
                #print(model_name)
            elif row.find("td", string = "Display Size"):
                display_size = row.find("li", class_="_21lJbe").text
                #print(display_size)
            elif row.find("td", string = "Operating System"):
                os = row.find("li", class_="_21lJbe").text
                #print(os)
            elif row.find("td", string = "HD Technology & Resolution"):
                pitcure_quality = row.find("li", class_="_21lJbe").text
                #print(pitcure_quality)
            elif row.find("td", string = "Smart Tv"):
                Smart_Tv = row.find("li", class_="_21lJbe").text
                #print(Smart_Tv)
            elif row.find("td", string = "Sound Technology"):
                Sound_Technology = row.find("li", class_="_21lJbe").text

                #print(Sound_Technology)
            elif row.find("td", string = "Refresh Rate"):
                Refresh_Rate = row.find("li", class_="_21lJbe").text
                #print(Refresh_Rate)
            elif row.find("td", string = "Screen Type"):
                Screen_Type = row.find("li", class_="_21lJbe").text
                #print(Screen_Type)
            elif row.find("td", string = "HDMI"):
                HDMI = row.find("li", class_="_21lJbe").text
                if not HDMI.__contains__("Side"):
                    hdmi = HDMI
                #print(HDMI)
            elif row.find("td", string = "USB"):
                USB = row.find("li", class_="_21lJbe").text
                if not USB.__contains__("Side"):
                    usb = USB
                #print(USB)
                connectivity = hdmi+" HDMI | "+ usb +" USB"
                #print(connectivity)
        price = soup.find("div", class_="_30jeq3 _16Jk6d").text
        #print(price)
        mob ={'Brand':brand,
             'Model Name': model_name,
             'Display Size': display_size,
             'Operating System': os,
             'Picture Quality': pitcure_quality,
             'Smart Tv': Smart_Tv,
             'Sound Technology': Sound_Technology,
             "Refresh Rate": Refresh_Rate,
             'Screen Type': Screen_Type,
             'Connectivity': connectivity,
             'Price': price}
        print(mob)
        tv.append(mob)
    except:
        rem_tv.append(link)
```
