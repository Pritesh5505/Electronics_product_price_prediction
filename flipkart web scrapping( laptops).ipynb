```python
import pandas as pd
import numpy as np 
import requests
import re
import time
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

```python
ua = UserAgent()
base_url = "https://www.flipkart.com"
```

```python
headers ={'User-Agent': ua.random}
r = requests.get("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=40",
                headers =headers)
print(r)
```

```python
remainig_page = []
product_links = []
for k in range(1,42):
    time.sleep(np.random.randint(2,5))
    try:
        headers ={'User-Agent': ua.random}
        r = requests.get(f"https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={k}",headers = headers)
        soup = BeautifulSoup(r.content,'lxml')
        product_list = soup.find_all('div', class_="_1YokD2 _3Mn1Gg")
        #print(product_list)
        
        for item in product_list:
            for link in item.find_all('div', class_ = "_1AtVbE col-12-12"):
                #print(link)
                for i in link.find_all('a', href = True):

                    product_links.append(base_url+i['href'])
        product_links = product_links[: len(product_links)-12]
    except :
        print('error has occured')
        remainig_page.append(k)
```

```python
remainig_products = []
laptops =[]
for link in product_links:
    headers ={'User-Agent': ua.random}
    try:
        
        r = requests.get(link,headers = headers)
        print(r)
        
        soup = BeautifulSoup(r.content,'lxml')
        name = soup.find('span', class_="B_NuCI").text
        #print(name)
        soup1 = soup.find_all('div', class_ = '_3k-BhJ')
        model_name = soup1[0].find_all('li',class_ = "_21lJbe")[3].text
        #print(model_name)

            Specificatios = soup.find_all("tr", class_="_1s_Smc row")
        for row in Specificatios:
            if row.find("td", string="Processor Name"):
                processor_name = row.find("li", class_="_21lJbe").text.strip()
            if row.find('td', string="Operating System"):
                operating_system = row.find("li", class_="_21lJbe").text.strip()
            if row.find('td', string=re.compile(r"\bCapacity\b", flags=re.IGNORECASE)):
                Storage = row.find("li", class_="_21lJbe").text.strip()
            if row.find('td', string="RAM"):
                RAM = row.find("li", class_="_21lJbe").text.strip()
            if row.find('td', string="Screen Size"):
                Screen_Size = row.find("li", class_="_21lJbe").text.strip() 
        Price = soup.find('div', class_="_30jeq3 _16Jk6d").text
        touch_screen = soup1[4].find_all('li',class_ = "_21lJbe")[0].text
        lap= {'Brand': name,
             'Model Name': model_name,
             'Processor': processor_name,
             'Operating System': operating_system,
             'Storage': Storage,
             'RAM': RAM,
             'Screen Size': Screen_Size,
             'Touch_Screen': touch_screen,
             'Price': Price}
        print(lap)
        laptops.append(lap)
        
    except:
        remainig_products.append(link)
    time.sleep(np.random.randint(1,3))
```
