```python
def remove_brackets(text):

    return re.sub(r'\([^)]*\)', '', text)

df['Brand'] = df['Brand'].apply(remove_brackets)
```

```python
df['Brand'] = df['Brand'].str.split(" ").str.slice(0,3).str.join(" ")
```

```python
df['Model Name'] = df['Model Name'].str.split(" ").str.slice(0,3).str.join(" ")
df.sample(5)
```

```python
df['Display Size'].unique()
```

```python
def display(data):
    if 'mm' in data.lower():
        return float(data.split(" ")[0])
    elif 'cm' in data.lower():
        return float(data.split(" ")[0])*10
    elif 'inch' in data.lower():
        return float(data.split(" ")[0])*25.4
    else:
        return None
```

```python
df['Display Size'] = df['Display Size'].apply(display)
```

```python
df.sample(5)
```

```python
df['Operating System'][df['Operating System'] == "No"] = np.nan
```

```python
df['Bluetooth'][df['Bluetooth'] == "Bluetooth"] = "Yes"
```

```python
def clean_display_type(display_type):
    display_type = str(display_type).lower()

    if 'super amoled' in display_type:
        return 'Super AMOLED'
    elif 'amoled' in display_type:
        return 'AMOLED'
    elif 'ips' and 'tft' in display_type:
        return "IPS TFt"
    elif 'ips' and 'lcd' in display_type:
        return 'IPS LCD'
    elif 'tft' and 'led' in display_type:
        return 'TFT LED'
    elif 'oled' and 'retina' in display_type:
        return 'OLED Retina'
    elif 'ips' in display_type:
        return 'IPS'
    elif 'tft' in display_type:
        return 'TFT'
    elif 'oled' in display_type:
        return 'OLED'
    elif 'lcd' in display_type:
        return 'LCD'
    elif 'hd' in display_type:
        return 'HD'
    else:
        return 'Other'
```

```python
df['Display_Type'] = df['Display_Type'].apply(clean_display_type)
df.sample(5)
```

```python
def clean_pixel_info(pixel_info):
    pixel_values = re.findall(r'\b\d+\b', str(pixel_info))
    if pixel_values:
        average_pixel_value = int(np.mean([int(value) for value in pixel_values]))
        return average_pixel_value
    else:
        return np.nan
```

```python
df.Display_Resolution =df.Display_Resolution.apply(clean_pixel_info)
df.sample(5)
```

```python
df.Touchscreen[df.Touchscreen == 'Touchscreen'] = 'Yes'
```

```python
df.Water_Resistance.unique()
```

```python
def clean_water_resis(data):
    if "mm" in data.lower():
        return float(data.split(" ")[0])/1000
    elif 'm' in data.lower():
        return float(data.split(" ")[0])
    else:
        return np.nan
```

```python
df.Water_Resistance = df.Water_Resistance.apply(clean_water_resis)
```

```python
def extract_days(text):
    matches = re.findall(r"\d+", text)
    if matches:
        durations = text.split(",")
        days = []
        for duration in durations:            
            num = float(re.findall(r"\d+", duration)[0])
            if any(s in duration for s in ["h", "hrs", "hours", "hr"]):
                num /= 24
            days.append(num)
        # Return the longest duration (assuming main battery life)
        return max(days)
    else:
        return np.nan
df.Battery_Life = df.Battery_Life.apply(extract_days)
```
