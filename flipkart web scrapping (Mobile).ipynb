```python
import requests
from bs4 import BeautifulSoup
base_url = "https://www.flipkart.com"
import pandas as pd 
import numpy as np
import time 
import re
from fake_useragent import UserAgent
```

```python
ua = UserAgent()
header = {'user-Agent' : ua.random}
```

```python
productlinks = []

for i in range(1,42):
    try:
        r = requests.get(f'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}',headers = header)
        #print(i)
        soup = BeautifulSoup(r.content, 'lxml')
        productlist = soup.find('div',class_="_1YokD2 _3Mn1Gg")
        #print(productlist)
        for item in productlist.find_all('div', class_= "_1AtVbE col-12-12"):
            for link in item.find_all('a',href = True):
                productlinks.append(base_url + link['href'])
        #len(productlinks)
        k=12
        productlinks = productlinks[: len(productlinks)-k]
        #len(productlinks)
    except Exception as e:
        print(i)
        print("error has occured")
    time.sleep(np.random.randint(1,3))
```

```python
mobile = []
rem_mob = []
for link in productlinks:
    try:
        r = requests.get(link,headers= header)
        print(r)
        soup = BeautifulSoup(r.content,'lxml')
        brand = soup.find('span', class_ = "B_NuCI").text
        soup1 = soup.find_all("tr", class_="_1s_Smc row")
        #print(soup1)
        model_name = display_size = os = processor = storage = pri_cam = sec_cam = network = battery = RAM = "NA"
        for row in soup1:
            if row.find("td", string = "Model Name"):
                model_name = row.find('li', class_ = "_21lJbe").text
                #print(model_name)
            elif row.find("td", string = "Display Size"):
                display_size = row.find("li", class_="_21lJbe").text
                #print(display_size)
            elif row.find("td", string = "Operating System"):
                os = row.find("li", class_="_21lJbe").text
                #print(os)
            elif row.find("td", string = "Processor Type"):
                processor = row.find("li", class_="_21lJbe").text
                #print(processor)
            elif row.find("td", string = "Internal Storage"):
                storage = row.find("li", class_="_21lJbe").text
                #print(storage)
            elif row.find("td", string = "RAM"):
                RAM_tag = row.find("li", class_="_21lJbe")
                RAM = RAM_tag.text if RAM_tag else "NA"
                #print(RAM)
            elif row.find("td", string = "Primary Camera"):
                pri_cam = row.find("li", class_="_21lJbe").text
                if pri_cam.__contains__("MP"): # since pri_cam running twice and 2nd iteration changing the desired value.
                    primary_cam = pri_cam
                #print(primary_cam)
            elif row.find("td", string = "Secondary Camera"):
                sec_cam = row.find("li", class_="_21lJbe").text
                #print(sec_cam)
            elif row.find("td", string = "Network Type"):
                network = row.find("li", class_="_21lJbe").text
                #print(network)
            elif row.find("td", string = "Battery Capacity"):
                battery = row.find("li", class_="_21lJbe").text
                #print(battery)
        price = soup.find("div", class_="_30jeq3 _16Jk6d").text
        #print(price)
        mob ={'Brand':brand,
             'Model Name': model_name,
             'Display Size': display_size,
             'Operating System': os,
             'Processor': processor,
             'Storage': storage,
             'RAM': RAM,
             "Primary camera": primary_cam,
             'Front Camera': sec_cam,
             'Network': network,
             'battery': battery,
             'Price': price}
        print(mob)
        mobile.append(mob)
    except:
        rem_mob.append(link)
```
